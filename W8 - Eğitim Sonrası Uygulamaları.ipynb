# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
class Kisi:
    def __init__(self, isim, soyisim, yas, boy, kilo):
        self.isim = isim
        self.soyisim = soyisim
        self.yas = yas
        self.boy = boy
        self.kilo = kilo
        self.vki = self.vkiHesapla()
        self.kategori = self.kategoriBelirle()

    def vkiHesapla(self):
        return self.kilo / (self.boy ** 2)

    def kategoriBelirle(self):
        if self.vki < 18.5:
            return "Zayıf"
        elif 18.5 <= self.vki < 24.9:
            return "Normal kilolu"
        elif 25 <= self.vki < 29.9:
            return "Fazla kilolu"
        elif 30 <= self.vki < 39.9:
            return "Obez"
        else:
            return "İleri derecede obez (morbid obez)"

def veriGirisi(dosyaAdi):
    try:
        girisSayisi = int(input("Kaç kişi için veri girişi yapmak istersiniz? "))
    except ValueError:
        print("Geçersiz giriş. Lütfen bir sayı girin.")
        return

    kisiler = []

    for _ in range(girisSayisi):
        isim = input("İsim: ")
        soyisim = input("Soyisim: ")
        try:
            yas = int(input("Yaş: "))
            boy = float(input("Boy (metre cinsinden): "))
            kilo = float(input("Kilo (kg cinsinden): "))
        except ValueError:
            print("Geçersiz giriş. Lütfen yaş, boy ve kilo için geçerli sayılar girin.")
            continue

        kisi = Kisi(isim, soyisim, yas, boy, kilo)
        kisiler.append(kisi)

    with open(dosyaAdi, 'w', encoding='utf-8') as dosya:
        for kisi in kisiler:
            dosya.write(f"{kisi.isim} {kisi.soyisim}: {kisi.kategori}\n")

    print("\nVKİ Kategorileri:")
    for kisi in kisiler:
        print(f"{kisi.isim} {kisi.soyisim}: {kisi.kategori}")

veriGirisi('kisiler.txt')


```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
import datetime

menu = {
    'Mango Smoothie': 12.0,
    'Berry Blast': 10.0,
    'Green Delight': 11.0,
    'Tropical Mix': 13.0
}

siparisler = []

def MenuGoruntule():
    print("\n--- Menü ---")
    for icerik, fiyat in menu.items():
        print(f"{icerik}: ${fiyat:.2f}")

def siparişler():
    MenuGoruntule()
    secim = input("\nHangi içeceği almak istersiniz? (İçecek adını yazın): ").strip()
    if secim in menu:
        fiyat = menu[secim]
        tarih = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        siparisler.append({'İçecek': secim, 'Fiyat': fiyat, 'Tarih': tarih})
        print(f"\nSiparişiniz alındı: {secim} - ${fiyat:.2f}")
        print(f"Tarih: {tarih}")
    else:
        print("Geçersiz içecek adı. Lütfen menüdeki içeceklerden birini seçin.")

def raporAl():
    print("\n--- Sipariş Raporu ---")
    if siparisler:
        for siparis in siparisler:
            print(f"İçecek: {siparis['İçecek']}, Fiyat: ${siparis['Fiyat']:.2f}, Tarih: {siparis['Tarih']}")
    else:
        print("Henüz sipariş verilmedi.")

def anaMenu():
    while True:
        print("\n--- YetBar - Smoothie Bank ---")
        print("1. Menüyü Görüntüle")
        print("2. Sipariş Ver")
        print("3. Sipariş Raporu")
        print("4. Çıkış")
        
        secim = input("Bir seçenek girin (1/2/3/4): ").strip()
        
        if secim == '1':
            MenuGoruntule()
        elif secim == '2':
            siparişler()
        elif secim == '3':
            raporAl()
        elif secim == '4':
            print("Programdan çıkılıyor...")
            break
        else:
            print("Geçersiz seçim. Lütfen 1, 2, 3 veya 4 girin.")

anaMenu()
```
