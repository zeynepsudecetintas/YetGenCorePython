# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
def urunlerDosyasi(urunler):
    urunAd = input("Ürün adını girin: ")
    urunFiyat = input("Ürün fiyatını girin: ")

    with open(urunler, 'a') as file:
        file.write(f"{urunAd}, {urunFiyat}\n")
    
    print(f"Ürün bilgisi {urunler} dosyasına kaydedildi: {urunAd}, {urunFiyat}")

urunlerDosyasi('urunler.txt')

````

# Problem 2
1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız 
ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python
def kelimeDegistir(dosyaAdi, eskiKelime, yeniKelime):
    with open(dosyaAdi, 'r', encoding='utf-8') as dosya:
        icerik = dosya.read()
    
    icerik = icerik.replace(eskiKelime, yeniKelime)
   
    with open(dosyaAdi, 'w', encoding='utf-8') as dosya:
        dosya.write(icerik)
    
    print(f"'{eskiKelime}' kelimesi '{yeniKelime}' kelimesi ile değiştirildi ve '{dosyaAdi}' dosyasına kaydedildi.")

dosyaAdi = input("Dosya adını girin: ")
eskiKelime = input("Değiştirilmesini istediğiniz eski kelimeyi girin: ")
yeniKelime = input("Yeni kelimeyi girin: ")

kelimeDegistir(dosyaAdi, eskiKelime, yeniKelime)
````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
class arac:
    def __init__(self, ad, hız, kilometre):
        self.ad = ad
        self.hız = hız
        self.kilometre = kilometre
    
    def bilgi(self):
        return f"Ad: {self.ad} Hız: {self.hız} Km: {self.kilometre}"

class otobus(arac):
    def __init__(self, ad, hız, kilometre):
        super().__init__(ad, hız, kilometre)

okulOtobusu = otobus("Okul Volvo", 180, 12)
print(okulOtobusu.bilgi())
````
